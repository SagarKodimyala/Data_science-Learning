# Categorical Data

```python
#countplot
sns.countplot(x="Survived", data=df)  #Graphical reprenstation for analysis. 1 = survived and 0 = not survived
df["Survived"].value_counts()
```

```python
sns.countplot(x="Pclass", data=df)
df["Pclass"].value_counts()
```

```python
sns.countplot(x= "Age", data=df)
df["Age"].value_counts()
```

```python
sns.countplot(x="Embarked", data=df)
df["Embarked"].value_counts()
```
